```python
from transformers import TFBertForMaskedLM ### ... 이 부분을 완성하시오
from transformers import AutoTokenizer ### ... 이 부분을 완성하시오
```

```python
model = TFBertForMaskedLM.from_pretrained('bert-large-uncased') ### ... 이 부분을 완성하시오
tokenizer = AutoTokenizer.from_pretrained('bert-large-uncased') ### ... 이 부분을 완성하시오
```

```python
inputs = tokenizer('Soccer is a really fun [MASK].', return_tensors='tf') # MLM 예측 결과
```

```python
print(inputs['input_ids']) # 정수 인코딩 결과
```

```python
print(inputs['token_type_ids']) # 세그먼트 인코딩 결과
```

```python
print(inputs['attention_mask']) # 어텐션 마스크 결과
```

```python
from transformers import FillMaskPipeline
pip = FillMaskPipeline(model=model, tokenizer=tokenizer) ### ... 이 부분을 완성하시오
```

```python
pip('Soccer is a really fun [MASK].')
```

```python
pip('The Avengers is a really fun [MASK].')
```
