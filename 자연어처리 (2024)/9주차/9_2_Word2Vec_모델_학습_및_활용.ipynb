````python
from gensim.models import Word2Vec  ### ... 이 부분을 완성하시오 (강의 자료를 참고하여 '???' 부분을 작성해 보시기 바랍니다.)
from konlpy.tag import Komoran  ### ... 이 부분을 완성하시오 (강의 자료를 참고하여 '???' 부분을 작성해 보시기 바랍니다.)
import time
````

## 파일 읽기 함수 작성

````python
def read_review(filename):
    f = open(filename, 'r', encoding = 'utf-8')
    data = []
    for line in f.readlines():
        if line != '':
            data.append(line.split('\t'))
    data = data[1:]
    f.close()

    return data
````

````python
data = read_review('ratings.txt')[:20000]
len(data)
````

## 형태소 분석

````python
# 1분 내외 소요
print('POS tagging...')
start = time.time()
komoran = Komoran()
docs = []
count = 0
for sent in data:
    if count % 1000 == 0:
        print(count, time.time()-start)
    docs.append(komoran.morphs(sent[1]))
    count += 1
print(docs[:10])
print('Done...', time.time()-start)
````

## Word2Vec 학습

````python
# 30초 내외 소요
print('Word2vec training...')
start = time.time()
model = Word2Vec(sentences=docs, vector_size=200, window=4, hs=1, min_count=5, sg=1, negative=0)  ### ... 이 부분을 완성하시오 (강의 자료를 참고하여 '???' 부분을 작성해 보시기 바랍니다.)
print('Done...', time.time()-start)
````

## 학습된 벡터 값 확인: Word2Vec

````python
print('영화 :', model.wv['영화'])
````

## 단어 유사도 계산: Word2Vec

````python
print(model.wv.most_similar('배우', topn=5))  ### ... 이 부분을 완성하시오 (강의 자료를 참고하여 '???' 부분을 작성해 보시기 바랍니다.)
````

````python
print(model.wv.most_similar('이정재', topn=5))
````

````python
print(model.wv.most_similar('감동', topn=5))
````

## 1. [fastText란?]
fastText가 무엇인지 설명하시오. (인터넷 검색 등을 통해 내용을 작성해 보시오.)
- 설명: fastText란 페이스북에서 개발한 Word Embedding 방법으로 Word2Vec 방식에서 철자 기반의 임베딩 방식을 추가한 새로운 임베딩 방식입니다. Word2Vec와 FastText와의 가장 큰 차이점이라면 Word2Vec는 단어를 쪼개질 수 없는 단위로 생각한다면, FastText는 하나의 단어 안에도 여러 단어들이 존재하는 것으로 간주합니다. 즉, FastText는 서브워드(subword)를 고려하여 학습합니다. 이를 통해 Word2Vec의 한계였던  OOV(Out of Vocabulary) 문제, 즉 학습된 말뭉치에 존재하지 않은 단어들을 벡터화할 수 없는 문제을 보완할 수 있습니다.

* 설명:

    GloVe는 카운트 기반의 LSA와 예측 기반의 Word2Vec의 단점을 보완하기 위해, 카운트 기반과 예측 기반을 결합한 방법론입니다. GloVe는 동시 등장 행렬과 동시 등장 확률을 활용해 손실 함수를 설계하고, 임베딩된 중심 단어와 주변 단어 벡터의 내적이 전체 코퍼스에서의 동시 등장 확률을 잘 근사하도록 학습합니다. 현재 연구에 따르면 GloVe와 Word2Vec 중 어느 것이 더 우수한지 단정하기 어렵기 때문에, 두 가지 방법을 모두 시도하여 더 성능이 좋은 것을 선택하는 것이 좋습니다.
