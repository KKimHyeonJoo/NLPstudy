# 파이프라인을 이용한 감성분석

```python
# 필요한 라이브러리를 설치하기
! pip install transformers ### ... 이 부분을 완성하시오
```

```python
# 파이프라인 라이브러리 불러오기
from transformers import pipeline ### ... 이 부분을 완성하시오

# 파이프라인 라이브러리 호출하기: 감성분석
sentiment_clf = pipeline("sentiment-analysis") ### ... 이 부분을 완성하시오

# "what a beautiful day!"라는 문장의 감성 확인하기
result = sentiment_clf("what a beautiful day!") ### ... 이 부분을 완성하시오

print(result)
print("감성분석 결과:", result[0]['label'])
print("감성 스코어:", result[0]['score'])
```

```python
from transformers import pipeline
# 오토 토크나이저, 오토 모델(시퀀스 분류) 불러오기
from transformers import AutoTokenizer, AutoModelForSequenceClassification ### ... 이 부분을 완성하시오

tokenizer = AutoTokenizer.from_pretrained("monologg/koelectra-base-finetuned-sentiment") ### ... 이 부분을 완성하시오
model = AutoModelForSequenceClassification.from_pretrained("monologg/koelectra-base-finetuned-sentiment") ### ... 이 부분을 완성하시오

sentiment_clf = pipeline("sentiment-analysis", tokenizer=tokenizer, model=model) ### ... 이 부분을 완성하시오

result = sentiment_clf("날이 참 좋아요!")

print(result)
print("감성분석 결과:", result[0]['label'])
print("감성 스코어:", result[0]['score'])
```

# 파이프라인을 이용한 질의 응답

```python
context = '''Text mining, also referred to as text data mining (abbr.: TDM),
        similar to text analytics,
        is the process of deriving high-quality information from text. It involves
        "the discovery by computer of new, previously unknown information,
        by automatically extracting information from different written resources."
        Written resources may include websites, books, emails, reviews, and articles.
        High-quality information is typically obtained by devising patterns and trends
        by means such as statistical pattern learning. According to Hotho et al.(2005)
        we can distinguish between three different perspectives of text mining:
        information extraction, data mining, and a KDD (Knowledge Discovery in Databas
        es) process.'''

question = "What is text mining?"
```

```python
from transformers import pipeline

# 질의응답 파이프라인 호출하기
question_answerer = pipeline("question-answering") ### ... 이 부분을 완성하시오

# 질문과 문단 전달하기
answer = question_answerer(question=question, context=context) ### ... 이 부분을 완성하시오

print(answer)
```

```python
context = """수원 화성은 조선시대 화성유수부 시가지를 둘러싼 성곽이다.
1789년(정조 13) 수원을 팔달산 동쪽 아래로 옮기고,
1794년(정조 18) 축성을 시작해 1796년에 완성했다."""
context = context.strip().replace("\n","")

question = "수원 화성은 언제 완성되었는가?"
```

```python
from transformers import pipeline
# 오토 토크나이저, 오토 모델(질의응답) 불러오기
from transformers import ElectraTokenizer, ElectraForQuestionAnswering, pipeline  ### ... 이 부분을 완성하시오


tokenizer = ElectraTokenizer.from_pretrained("monologg/koelectra-small-v2-distilled-korquad-384") ### ... 이 부분을 완성하시오
model = ElectraForQuestionAnswering.from_pretrained("monologg/koelectra-small-v2-distilled-korquad-384") ### ... 이 부분을 완성하시오

question_answerer = pipeline("question-answering", tokenizer=tokenizer, model=model) ### ... 이 부분을 완성하시오
answer = question_answerer({"question": question, "context": context}) ### ... 이 부분을 완성하시오

print(answer)
```

## 1. 관심있는 키워드에 대해 검색 후 질의응답 해보는 코드 작성

wikipedia.search에 임의의 검색어를 입력 (Ex. '뉴진스')

```python
# 제목을 검색(search)하시오
wiki_title = wikipedia.search('두아 리파') # ('뉴진스') ### 이 부분을 새로운 검색어로 작성하시오
wiki_title
```

wikipedia.page는 위의 셀 wikipedia.search의 결과로 나온 것 중 하나를 선택 (Ex. NewJeans)
- ['윤성호 (희극인)',
 '더 팩트 뮤직 어워즈',
 '최현욱',
 '케플러 (음악 그룹)',
 **'NewJeans'**,
 '박지후',
 'LG PC 그램',
 '김민지',
 '단월드',
 '롤라팔루자']
 - 가장 적합한 'NewJeans' 선택

```python
# 위키피디아 페이지(page)를 가져오시오
wiki_text = wikipedia.page('Future Nostalgia').content # ('NewJeans').content ### 이 부분을 새롭게 검색된 결과로 작성하시오
wiki_text
```

```python
context = wiki_text
context = context.strip().replace("\n","")
question = "두아리파는 어디 나라 가수야?" # "뉴진스는 몇 인조 그룹이야?" ### 이 부분을 새로운 질문으로 작성하시오
```

```python
from transformers import ElectraTokenizer, ElectraForQuestionAnswering, pipeline

tokenizer = ElectraTokenizer.from_pretrained("monologg/koelectra-small-v2-distilled-korquad-384")
model = ElectraForQuestionAnswering.from_pretrained("monologg/koelectra-small-v2-distilled-korquad-384")

question_answerer = pipeline("question-answering", tokenizer=tokenizer, model=model)
answer = question_answerer({"question": question, "context": context})

print(answer)
```
